```python
from PIL import ImageFilter

from pattern import Caretaker, ImageEditor
```

```python
img_editor = ImageEditor('./avatar.png')
careteker = Caretaker(img_editor)
```

```python
img_editor.show()
```

```python
careteker.backup()
```

```python
img_editor.add_filter(ImageFilter.CONTOUR)
```

```python
img_editor.show()
```

```python
careteker.undo()
img_editor.show()
```

```python
careteker.backup()
img_editor.add_filter(ImageFilter.BLUR)
```

```python
img_editor.show()
```

```python
careteker.backup()
img_editor.to_gray_scale()
```

```python
img_editor.show()
```

```python
img_editor.to_file('avatar-filter.png')
```

```python
careteker.show_history()
```

```python
careteker.undo()
careteker.undo()

img_editor.show()
```
